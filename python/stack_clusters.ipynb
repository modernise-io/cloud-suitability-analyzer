```python
import pandas as pd
from sklearn.cluster import OPTICS
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import silhouette_score

# Load your data
file_path = 'data/pivot_stacks.csv'
data = pd.read_csv(file_path)

# Handle missing values
data = data.fillna("None")

# Convert categorical variables to numerical format
label_encoder = LabelEncoder()
data_encoded = data.apply(label_encoder.fit_transform)

# Perform OPTICS clustering
optics = OPTICS(min_samples=2)
labels = optics.fit_predict(data_encoded)

# Assign clusters to the original data
data['cluster'] = labels

# Evaluate the clusters
silhouette_avg = silhouette_score(data_encoded, labels)
print("Silhouette Score: ", silhouette_avg)

# Move the cluster column to the first position
data = data[['cluster'] + [col for col in data.columns if col != 'cluster']]

# Sort the data by cluster
data = data.sort_values(by='cluster')

# Save the result
data.to_csv('data/clustered_data.csv', index=False)

# Print the result
print(data)
```
